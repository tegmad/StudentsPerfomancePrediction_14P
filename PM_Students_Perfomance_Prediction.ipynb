# üìÑ Dataset Description: Student Academic Performance

This dataset contains demographic, behavioral, and academic details of students, intended for analysis and classification based on academic performance.

---

## üÜî Student Identification
- **StudentID**: A unique identifier assigned to each student (ranging from **1001 to 3392**).

---

## üë• Demographic Details
- **Age**: Student age, ranging from **15 to 18 years**.
- **Gender**: 
  - `0` ‚Äî Male  
  - `1` ‚Äî Female
- **Ethnicity**: 
  - `0` ‚Äî Caucasian  
  - `1` ‚Äî African American  
  - `2` ‚Äî Asian  
  - `3` ‚Äî Other
- **ParentalEducation**: Education level of parents:
  - `0` ‚Äî None  
  - `1` ‚Äî High School  
  - `2` ‚Äî Some College  
  - `3` ‚Äî Bachelor's  
  - `4` ‚Äî Higher

---

## üìö Study Habits
- **StudyTimeWeekly**: Weekly study time in hours (**0 to 20**).
- **Absences**: Number of school absences (**0 to 30**).
- **Tutoring**: 
  - `0` ‚Äî No  
  - `1` ‚Äî Yes

---

## üë®‚Äçüë©‚Äçüëß Parental Involvement
- **ParentalSupport**: Level of parental support:
  - `0` ‚Äî None  
  - `1` ‚Äî Low  
  - `2` ‚Äî Moderate  
  - `3` ‚Äî High  
  - `4` ‚Äî Very High

---

## üé≠ Extracurricular Activities
- **Extracurricular**: Participation in extracurriculars:
  - `0` ‚Äî No  
  - `1` ‚Äî Yes
- **Sports**: Participation in sports:
  - `0` ‚Äî No  
  - `1` ‚Äî Yes
- **Music**: Participation in music activities:
  - `0` ‚Äî No  
  - `1` ‚Äî Yes
- **Volunteering**: Participation in volunteering:
  - `0` ‚Äî No  
  - `1` ‚Äî Yes

---

## üéì Academic Performance
- **GPA**: Grade Point Average (scale **2.0 to 4.0**), influenced by study habits, parental support, and extracurricular involvement.

### üéØ Target Variable ‚Äî GradeClass
Classification of students based on their **GPA**:
- `0` ‚Äî Grade **A** (GPA ‚â• 3.5)  
- `1` ‚Äî Grade **B** (3.0 ‚â§ GPA < 3.5)  
- `2` ‚Äî Grade **C** (2.5 ‚â§ GPA < 3.0)  
- `3` ‚Äî Grade **D** (2.0 ‚â§ GPA < 2.5)  
- `4` ‚Äî Grade **F** (GPA < 2.0)

## Task 1: Seeing the proportion of Data in the Data set


```python
df_by_gender = df_students.groupby('Gender')['Gender'].count().reset_index(name='count')
df_by_gender['Gender'] = df_by_gender['Gender'].map({0:'Male', 1:'Female'})


fig = px.pie(
    df_by_gender,
    names='Gender',
    values='count',
    color='Gender',
    color_discrete_map={'Male': 'blue', 'Female': 'deeppink'},
    title='Gender Distribution'
)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(
    width = 800,
    height = 600,
    showlegend=True)
fig.show()
```

```python
df_by_gender.style.background_gradient()
```

```python
def gender(x):
    if x == 0:
        return 'Male'
    else:
        return 'Female'
df_students['Gender_'] = df_students['Gender'].apply(gender)
```

```python
import plotly.express as px

fig = px.histogram(
    df_students,
    x='Age',                  
    color='Gender_',            
    nbins=20,                  
    title='Age Distribution of Students by Gender',
    color_discrete_map={'Male': 'blue', 'Female': 'deeppink'},  
    barmode='group',           
    text_auto=True             
)

fig.update_layout(
    width=1000,
    height=600,
    xaxis_title='Age',
    yaxis_title='Number of Students',
    bargap=0.1
)

fig.show()
```

```python
# Destribution by Etnicity

df_ethnicity = df_students[['Ethnicity','Age','Gender','ParentalEducation','Absences','GPA','Volunteering']]
df_ethnicity['Ethnicity'] = df_ethnicity['Ethnicity'].map({0:'Caucasian',1:'African American',2:'Asian',3:'Other'})
df_ethnicity.head()
```

```python
df_by_ethnicity = df_ethnicity.groupby(['Ethnicity','Age'])[['Gender']].count().reset_index()
df_by_ethnicity = df_by_ethnicity.rename(columns={'Gender':'Count'})
df_by_ethnicity.sort_values(by='Ethnicity').style.background_gradient(subset=['Count'],cmap='Reds')
```

```python
df_by_ethnicity_avg_GPA = df_ethnicity.groupby(['Ethnicity','Age'])['GPA'].mean().reset_index(name='Avg_GPA')
df_combined = df_by_ethnicity.merge(df_by_ethnicity_avg_GPA,on=['Ethnicity','Age'])

df_combined.style.background_gradient(subset=['Count','Avg_GPA'],cmap='Blues')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_ethnicity_2 = df_students.copy()
df_ethnicity_2['Ethnicity name'] = df_ethnicity_2['Ethnicity'].map({
    0: 'Caucasian', 1: 'African American', 2: 'Asian', 3: 'Other'
})
df_ethnicity_2 = df_ethnicity_2.groupby(
    ['Ethnicity name', 'Gender_', 'GradeClass']
)['Age'].count().reset_index()
df_ethnicity_2 = df_ethnicity_2.rename(columns={'Age': 'Count'})
df_ethnicity_2['Grade Explanation'] = df_ethnicity_2['GradeClass'].map({
    0: '>=3.5', 1: '3 ~ 3.5', 2: '2.5 ~ 3.0', 3: '2 ~ 2.5', 4: '<=2.0'
})

ethnicity_groups = ['African American', 'Asian', 'Caucasian', 'Other']
colors = {'Male': 'blue', 'Female': 'deeppink'}

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()  

for i, ethnicity in enumerate(ethnicity_groups):
    ax = axes[i]
    subset = df_ethnicity_2[df_ethnicity_2['Ethnicity name'] == ethnicity]

    barplot = sns.barplot(
        x='Grade Explanation',
        y='Count',
        hue='Gender_',
        data=subset,
        palette=colors,
        ax=ax
    )

    ax.set_title(f'{ethnicity} Students by Grade and Gender')
    ax.set_xlabel('Grade')
    ax.set_ylabel('Count')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
    ax.legend(title='Gender')

    for container in barplot.containers:
        barplot.bar_label(container, fmt='%d', label_type='edge', fontsize=9, padding=2)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,6))

male_df = df_students[df_students['Gender_'] == 'Male']
female_df = df_students[df_students['Gender_'] == 'Female']

sns.histplot(x='GPA', data=female_df, kde=True, color='deeppink', label='Female')
sns.histplot(x='GPA', data=male_df, kde=True, color='blue', label='Male')

female_mean = female_df['GPA'].mean()
female_median = female_df['GPA'].median()
male_mean = male_df['GPA'].mean()
male_median = male_df['GPA'].median()

plt.axvline(female_mean, color='deeppink', linestyle='--', linewidth=2, 
            label=f'Female Mean: {female_mean:.2f}')
plt.axvline(female_median, color='deeppink', linestyle=':', linewidth=2, 
            label=f'Female Median: {female_median:.2f}')
plt.axvline(male_mean, color='blue', linestyle='--', linewidth=2, 
            label=f'Male Mean: {male_mean:.2f}')
plt.axvline(male_median, color='blue', linestyle=':', linewidth=2, 
            label=f'Male Median: {male_median:.2f}')

plt.legend()
plt.xlim()
plt.tight_layout()
plt.show()
```
